```
csv_dir = "/content/drive/My Drive/CS4602/assignment1/csv/"
splittedByCountry_dir = csv_dir + "splittedByCountry/"
# raw_data = csv_dir + "COVID-19.csv"
raw_data = 'https://opendata.ecdc.europa.eu/covid19/casedistribution/csv'
```

```
import pandas as pd
import numpy as np
```

```
df = pd.read_csv(raw_data)
df.head(5)
```

```
header_to_kept = ['datetime', 'popData2019', 'deaths', 'cases']
```

```
df = df[df['year'] >= 2020].drop(["dateRep", "geoId", "countryterritoryCode", "continentExp", "Cumulative_number_for_14_days_of_COVID-19_cases_per_100000"], axis=1)
```

```
name_of_countries = df['countriesAndTerritories'].unique()
```

```
for country in name_of_countries:
    
    # print('Start creating ' + country + '.csv')
    splitted_data = splittedByCountry_dir + country + ".csv"

    df_country = df[df['countriesAndTerritories'] == country].copy()
    df_country = df_country.drop(['countriesAndTerritories'], axis=1)
    df_country['datetime'] = pd.to_datetime(df_country[['year', 'month', 'day']])
    df_country = df_country.sort_values(by=['datetime'], ascending=True)
    df_country = df_country.replace(np.nan, 0)
    df_country[['popData2019', 'deaths', 'cases']] = df_country[['popData2019', 'deaths', 'cases']].abs()
    df_country = df_country[header_to_kept]
    df_country = df_country.reset_index(drop=True)

    # print(df_country.tail(5))
    df_country.to_csv(splitted_data, header=header_to_kept, index=False)
  
  # print("Finish:", country)
```
